## Exercﾃｭcio 1

Considere uma tabela com as seguintes variﾃ｡veis:

| Variﾃ｡vel | Tipo |
|----------|------|
| nome_paﾃｭs | str |
| continente | str|
| area     | float|
| populacao | str  |
| pib     | float|


Um paﾃｭs ﾃｩ considerado grande se:
- sua ﾃ｡rea ﾃｩ de pelo menos 3 milhﾃｵes de km2 (3.000.000 km2), ou
- sua populaﾃｧﾃ｣o ﾃｩ de pelo menos 25 milhﾃｵes de pessoas.

Escreva uma soluﾃｧﾃ｣o que retorne o nome, a populaﾃｧﾃ｣o e a ﾃ｡rea dos paﾃｭses que sﾃ｣o considerados grandes. A saﾃｭda deve ser um dataframe com as colunas na ordem "ﾃ｡rea, nome, populaﾃｧﾃ｣o". 

```python
#str ﾃｩ string (texto; quali) e float ﾃｩ nﾃｺmero com casas decimais (quant)
 #a diferenﾃｧa de int para float ﾃｩ que, mesmo que ambos sejam nﾃｺmeros, int nﾃ｣o tem casas decimais
import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    # RESOLVA
    world['populacao'] = world['populacao'].astype(int)
    return world[(world['populacao'] >= 25 * 10**6) | (world['area'] >= 3 * 10**6)]
# world['populacao'] >= 25 * 10**6: seleciona as linhas onde a populaﾃｧﾃ｣o ﾃｩ maior ou igual a 25M (25*10竅ｶ=25000000)
# world['area'] >= 3 * 10**6: seleciona as linhas onde a ﾃ｡rea (coluna 'area') ﾃｩ maior ou igual a 3M
# .astype(int): forﾃｧa os valores da coluna populacao a serem tratados como nﾃｺmeros inteiros
# usei o |, que ﾃｩ um representa um OU lﾃｳgico. Portanto, o cﾃｳdigo filtra as linhas onde uma ou ambas as condiﾃｧﾃｵes (populaﾃｧﾃ｣o ou ﾃ｡rea) sﾃ｣o verdadeiras.

# Teste (Nﾃグ ALTERAR)
world = pd.DataFrame({
    'nome_paﾃｭs': ['Afeganistﾃ｣o', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'continente': ['ﾃ《ia', 'Europa', 'ﾃ’rica', 'Europa', 'ﾃ’rica'],
    'area': [652230, 28748, 2381741, 468, 1246700],
    'populacao': ['25500100', '2831741', '37100000', '78115', '20609294']
})

big_countries(world)
```

Nesses casos, o bﾃｴnus ﾃｩ de apenas 10%. Retorne um dataframe com as colunas `id_empregado` e `bonus`. Vocﾃｪ pode usar a biblioteca Numpy. 

```python
import numpy as np

def bonus(funcionarios: pd.DataFrame) -> pd.DataFrame:
    # RESOLVA
    condition = funcionarios['id_empregado'] % 2 == 1 & ~funcionarios['nome'].str.startswith('M')
    funcionarios['bonus'] = np.where(condition, funcionarios['salario'], funcionarios['salario'] * 0.1)
    return funcionarios
funcionarios = pd.DataFrame({
    'id_empregado': [1, 2, 3, 4, 5, 6],
    'nome': ['Matheus', 'Marcela', 'Catarina', 'Clarissa', 'Lindsey', 'Maria'],
    'salario': [90_000, 10_000, 9_500, 9_999, 8_998, 4_111]
})

bonus(funcionarios)
```
